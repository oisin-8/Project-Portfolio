```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read in the data
df = pd.read_csv("/Users/oisinkelly/Downloads/Movies.csv")
```

```python
df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types

print(df.dtypes)
```

```python
df['year_correct'] = df['released'].astype(str).str[:4]
```

```python
df
```

```python
# Checking Outliers

df.boxplot(column=['gross'])
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.head()
```

```python
df.drop_duplicates()
```

```python
#Scatter plot of Gross Revenue compared to budget
plt.title('Budget v Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Film Budget')
plt.scatter(x=df['budget'], y=df["gross"])
```

```python
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "salmon"}, line_kws={"color":"orange"})
```

```python
#Pearson Correlation Matrix
df.corr(method ='pearson')
```

```python
# There is a high correlation between the Film's Budget and the Gross Revenue
```

```python
correlation_matrix = df.corr()
plt.title("Pearson Correlation Matrix - Numeric Breakdown")

plt.xlabel("Movie Stats")

plt.ylabel("Movie Stats")

sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
plt.title("Pearson Correlation Matrix - Movies")
plt.xlabel("Movie Stats")
plt.ylabel("Movie Stats")
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
# Top 10 Companies per Gross Revenue
CompanyGrossSum = df.groupby('company')[["gross"]].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:10]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 
CompanyGrossSumSorted
```
